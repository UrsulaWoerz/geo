# Python üêç

```python
import geopandas
db = geopandas.read_file(
    (
        "http://d2ad6b4ur7yvpq.cloudfront.net/"\
         "naturalearth-3.3.0/ne_110m_land.geojson"
    )
)
db.plot();
```

```python
db.to_parquet("db.pq", compression=None)
```

# R ü¶ú

```python
%%R
library(sfarrow)

db <- st_read_parquet("db.pq")
plot(db$geometry)
```

```python
%%R
st_write_parquet(db, "db2.pq")
```
